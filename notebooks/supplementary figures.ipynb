# Supplementary Figures
This notebook reproduces Figure S1 (DCA on test set) and Figure S2 (alert rule persistence sensitivity).

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rcParams

rcParams['font.family'] = 'sans-serif'
rcParams['font.sans-serif'] = ['Helvetica', 'Arial']
rcParams['axes.linewidth'] = 1.5
rcParams['xtick.major.width'] = 1.5
rcParams['ytick.major.width'] = 1.5
rcParams['xtick.labelsize'] = 10
rcParams['ytick.labelsize'] = 10

color_trauma = '#E64B35'
color_gray = '#666666'
color_lead = '#E64B35'
color_fpr = '#4DBBD5'

data_dir = './data'
```

```python
# Figure S1: DCA on test set (25% prevalence)
df_s1 = pd.read_excel(f'{data_dir}/FigureS1_DCA_TestSet.xlsx', sheet_name='Sheet1')
threshold = df_s1['Threshold']
nb_trauma = df_s1['Net_Benefit_TraumaFormer']
nb_treat_all = df_s1['Net_Benefit_TreatAll']
nb_treat_none = df_s1['Net_Benefit_TreatNone']

fig_s1, ax_s1 = plt.subplots(figsize=(8, 6))
ax_s1.plot(threshold, nb_trauma, color=color_trauma, linewidth=2.5, label='Trauma-Former')
ax_s1.plot(threshold, nb_treat_all, color=color_gray, linestyle='--', linewidth=2, label='Treat all')
ax_s1.plot(threshold, nb_treat_none, color=color_gray, linestyle=':', linewidth=2, label='Treat none')
ax_s1.set_xlabel('Threshold Probability', fontsize=12)
ax_s1.set_ylabel('Net Benefit', fontsize=12)
ax_s1.set_xlim(0, 1)
y_min = min(nb_trauma.min(), nb_treat_all.min(), -0.05)
y_max = nb_trauma.max() * 1.1
ax_s1.set_ylim(y_min, y_max)
ax_s1.grid(True, linestyle='--', alpha=0.5)
ax_s1.legend(loc='upper right', fontsize=10, frameon=False)
ax_s1.spines['top'].set_visible(False)
ax_s1.spines['right'].set_visible(False)
plt.tight_layout()
plt.savefig('FigureS1_DCA_TestSet.tiff', dpi=300, format='tiff', bbox_inches='tight')
plt.show()
```

```python
# Figure S2: Persistence sensitivity
df_s2 = pd.read_excel(f'{data_dir}/FigureS2_Persistence_Sensitivity.xlsx', sheet_name='Sheet1')
persistence = df_s2['Persistence_min']
lead_time = df_s2['Lead_Time_min']
fpr = df_s2['FPR_percent']

fig_s2, ax1_s2 = plt.subplots(figsize=(8, 6))
ax1_s2.plot(persistence, lead_time, color=color_lead, marker='o', markersize=8,
            linewidth=2.5, label='Lead time')
ax1_s2.set_xlabel('Persistence requirement (min)', fontsize=12)
ax1_s2.set_ylabel('Early warning lead time (min)', fontsize=12, color=color_lead)
ax1_s2.tick_params(axis='y', labelcolor=color_lead)
ax1_s2.set_xlim(0.5, 5.5)
ax1_s2.set_xticks([1,2,3,4,5])
ax1_s2.grid(True, linestyle='--', alpha=0.5)

ax2_s2 = ax1_s2.twinx()
ax2_s2.plot(persistence, fpr, color=color_fpr, marker='s', markersize=8,
            linewidth=2.5, label='FPR')
ax2_s2.set_ylabel('False positive rate (%)', fontsize=12, color=color_fpr)
ax2_s2.tick_params(axis='y', labelcolor=color_fpr)

lines1, labels1 = ax1_s2.get_legend_handles_labels()
lines2, labels2 = ax2_s2.get_legend_handles_labels()
ax1_s2.legend(lines1 + lines2, labels1 + labels2, loc='upper right', fontsize=10, frameon=False)

ax1_s2.spines['top'].set_visible(False)
ax2_s2.spines['top'].set_visible(False)
plt.tight_layout()
plt.savefig('FigureS2_Persistence_Sensitivity.tiff', dpi=300, format='tiff', bbox_inches='tight')
plt.show()
```
